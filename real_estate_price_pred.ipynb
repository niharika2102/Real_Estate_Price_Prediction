```python
def is_float(x):
  try:
    float(x)
  except:
    return False
  return True
```

```python
def convert_sqft_to_num(x):
  tokens = x.split('-')
  if len(tokens)==2:
    return (float(tokens[0])+float(tokens[1]))/2
  try:
    return float(x)
  except:
    return None
```

```python
def remove_pps_outliers(df):
  df_out = pd.DataFrame()
  for key, subdf in df.groupby('location'):
    m = np.mean(subdf.price_per_sqft)
    st = np.std(subdf.price_per_sqft)
    reduced_df = subdf[(subdf.price_per_sqft>(m-st))&(subdf.price_per_sqft<=(m+st))]
    df_out = pd.concat([df_out,reduced_df],ignore_index = True)
  return df_out

df7 = remove_pps_outliers(df6)
df7.shape
```

```python
def plot_scatter_chart(df,location):
  bhk2 = df[(df.location == location)&(df.bhk==2)]
  bhk3 = df[(df.location == location)&(df.bhk==3)]
  matplotlib.rcParams['figure.figsize']=(15,10)
  plt.scatter(bhk2.total_sqft,bhk2.price,color='blue',label='2BHK',s=50)
  plt.scatter(bhk3.total_sqft,bhk3.price,marker='+',color='green',label='3 BHK',s=50)
  plt.xlabel("Total Square Foot Area")
  plt.ylabel("Price")
  plt.title(location)
  plt.legend()

plot_scatter_chart(df7,"Hebbal")
```

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
df8 = remove_bhk_outliers(df7)
# df8 = df7.copy()
df8.shape
```

Model Building

```python
#one hot encoding to convert the location text data into numerical for ml model

dummies = pd.get_dummies(df10.location)
dummies.head(3)
```

```python
#one hot encoding rule to avoid dummy trap is to drop one column since it will be 0 anyways when the others are 1

df11 = pd.concat([df10,dummies.drop('other',axis = 'columns')],axis = 'columns')
df11.head(3)
```

```python
#dropping the location column since all columns are one hot encoded
df12 = df11.drop('location', axis = 'columns')
df12.head(2)
```

```python
df12.shape
```

```python
#x contains only the independent variables
#dropping price since its a dependent variable
#x is the df for model training

x = df12.drop('price',axis='columns')
x.head()
```

```python
x.shape
```

```python
#y is the dependent variable containting price

y = df12.price
y.head()
```

```python
len(y)
```

```python
#splitting training and testing data set
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest, ytrain,ytest = train_test_split(x,y, test_size=0.2,random_state = 10)

#test size 0.2 means 20% of data to be test and 80% to be training
```

```python
#Linear Regression model

from sklearn.linear_model import LinearRegression
lrmodel = LinearRegression()
lrmodel.fit(xtrain,ytrain)
lrmodel.score(xtest,ytest)
```

```python
 #Shuffle Split will randomize the data to make the distribution equal for the k fold cross validation and not targeted into one area

 from sklearn.model_selection import ShuffleSplit
 from sklearn.model_selection import cross_val_score

 cv = ShuffleSplit(n_splits = 5, test_size = 0.2, random_state = 0)

 cross_val_score(LinearRegression(), x,y, cv=cv)
```

```python
 from sklearn.model_selection import GridSearchCV

 from sklearn.linear_model import Lasso
 from sklearn.tree import DecisionTreeRegressor

 def find_best_model_using_gridsearchcv(x,y):
  algos = {
      'linear_regression':{
          'model' : LinearRegression(),
          'params' :{            
              'normalize': [True,False]
          }
      },
      'lasso':{
          'model':Lasso(),
          'params':{
              'alpha':[1,2],
              'selection' : ['random','cyclic']
          }
      },
      'decision_tree':{
          'model':DecisionTreeRegressor(),
          'params':{
              'criterion': ['mse','friedman_mse'],
              'splitter':['best','random']
          }
      }
  }


  scores = []
  cv = ShuffleSplit(n_splits = 5,test_size = 0.2, random_state = 0) #randomly shuffles sample for better result
  for algo_name, config in algos.items():
    gs = GridSearchCV(config['model'],config['params'], cv = cv, return_train_score = False)
    gs.fit(x,y)
    scores.append({
        'model':algo_name,
        'best_score' : gs.best_score_,
        'best_params': gs.best_params_
    })
  return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(x,y)
```

```python
def predict_price(location,sqft,bath,bhk):  #input values
  loc_index = np.where(x.columns == location)[0][0] 

  X = np.zeros(len(x.columns))   #zeros array
  X[0] = sqft   #1st column
  X[1] = bath   #2nd column
  X[2] = bhk    #3rd column
  if loc_index >=0:
    X[loc_index] = 1
  return lrmodel.predict([X])[0]
```
